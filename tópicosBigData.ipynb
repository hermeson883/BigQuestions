```python
import pandas as pd
import numpy as np
import plotly.express as px
```

```python
dados = pd.read_excel('Vendas.xlsx')
dados
```

```python
px.histogram(y=dados['Valor Final'], x = dados['ID Loja'])
```

```python
agregate = pd.DataFrame.agg(dados['Valor Final'], 'sum')
agregate
```

```python
df = pd.DataFrame()
maior_venda = dados[['Valor Final', 'ID Loja']]
maior_venda.sort_values(by='Valor Final', ascending=False)
# menor para o maior
```

```python
maior_venda_unitaria = dados[['Valor Unitário', 'ID Loja']]
maior_venda_unitaria.sort_values(by='Valor Unitário', ascending=False)
```

```python
#Quantidade de produtos das lojas
dados[['Quantidade', 'ID Loja']].groupby('ID Loja').agg('sum')
```

```python
#Média de vendas de cada loja (Valor Final)
dados[['Valor Final', 'ID Loja']].groupby('ID Loja').agg('mean')
```

```python
#Média de vendas de cafa loja (Valor Unitário)
dados[['Valor Unitário', 'ID Loja']].groupby('ID Loja').agg('mean')
```
